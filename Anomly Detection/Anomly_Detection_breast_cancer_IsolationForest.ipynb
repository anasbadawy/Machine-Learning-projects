```python
import numpy as np
import pandas as pd
data = pd.read_csv('breast-cancer-unsupervised-ad.csv',header=None)
#data = pd.read_csv('kdd99-unsupervised-ad.csv',header=None)
```

```python
X = data.iloc[:,:30].values
#X = data.iloc[:,:29].values
y = data.iloc[:,-1]
y=[1 if x=='n' else -1 for x in y]
```

```python
from sklearn.ensemble import IsolationForest

isolation_forest = IsolationForest(behaviour='new', bootstrap=False, contamination='auto',
        max_features=30, max_samples='auto', n_estimators=100,
        n_jobs=None, random_state=None, verbose=0)

isolation_forest.fit(X)
anomaly_score = isolation_forest.decision_function(X)
outlier = isolation_forest.predict(X)
```

```python
# from sklearn.metrics import accuracy_score
# accuracy_score(outlier,y)
```

```python
import matplotlib.pyplot as plt

dataset = pd.DataFrame({'anomaly_score':anomaly_score,'outlier':outlier})
dataset.plot.scatter(x='anomaly_score' ,y='outlier',color='DarkBlue', label='Group 1');
```
