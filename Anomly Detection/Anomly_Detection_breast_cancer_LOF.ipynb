```python

#importing the necessary libraries
import pandas as pd
import numpy as np
import os
import sys
from pyod.models.lof import LOF


#Reading the Data using Pandas 

data=pd.read_csv('breast-cancer-unsupervised-ad.csv', header=None)
#data=pd.read_csv('kdd99-unsupervised-ad.csv', header=None)

#putting the first 30(Breast Cancer file), 29(KDD file) columns and then converting the last column in each to binary 

#X= data.iloc[:,:29]  # for the KDD data set
X= data.iloc[:,:30]
y=data.iloc[:,-1]
#make y binary
y=[1 if x=='o' else 0 for x in y]

```

```python
# Initializing the classifier then fitting the model using X as training data
clf_name = 'LOF'
clf = LOF()

outlier = clf.fit_predict(X)  #outlier labels (0 or 1)
anomaly_score = clf.decision_function(X)  # outlier scores

```

```python
#plotting the Outliers with anomaly score
import matplotlib.pyplot as plt

dataset = pd.DataFrame({'anomaly_score':anomaly_score,'outlier':outlier})
dataset.plot.scatter(x='anomaly_score' ,y='outlier',color='red', label='Data');
```
